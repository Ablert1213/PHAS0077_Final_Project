```python
import matplotlib.pyplot as plt
import re

```

```python
# Initialize lists to hold data
file_names = []
chi2_values = []
p_values = []
N_values = []

# Read the file
with open("chi_square_results_10_to_10.txt", "r") as f:
    for line_num, line in enumerate(f, 1):
        # Skip lines without data
        if "File" not in line:
            continue
        
        # Split line and extract data
        parts = line.split()

        file_names.append(parts[1][10:-1])  # remove the trailing colon
        chi2_values.append(float(parts[4][:-1]))  # remove the trailing comma
        p_values.append(float(parts[7][:-1]))  # remove the trailing comma
        N_values.append(int(parts[10]))


# Plotting
fig, ax1 = plt.subplots(figsize=(15, 7))

# Twin the axes for two different y-axes
ax2 = ax1.twinx()

# Plot with labels for legend
ax1.plot(file_names, chi2_values, 'g-', label='Chi^2')
ax2.plot(file_names, p_values, 'b-', label='p-value')
ax1.set_xticks(ax1.get_xticks()[::5])

# Set the y axis labels
ax1.set_ylabel('Chi^2', color='g')
ax2.set_ylabel('p-value', color='b')

# Add legends to the axes
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Rotate x labels for better visibility
plt.setp(ax1.get_xticklabels(), rotation=45, ha="right")

# Provide a title and show the plot
plt.title("Chi^2 and p-values for each file in 10 to 10")
plt.tight_layout()
plt.show()
```

```python

# Read data from the file
with open('MAD_results_10_to_10.txt', 'r') as file:
    data_a = file.read()

# Regular expression to extract MAD values and file names
pattern = r"File (.*?): MAD = (.*?),"
matches = re.findall(pattern, data_a)

file_names = [match[0] for match in matches]
mad_values = [float(match[1]) for match in matches]

# Plotting the MAD values with threshold lines
plt.figure(figsize=(15, 8))
plt.plot(file_names, mad_values, marker='o', linestyle='-', color='blue')
plt.axhline(y=0.006, color='green', linestyle='--', label="Close conformity")
plt.axhline(y=0.012, color='yellow', linestyle='--', label="Acceptable conformity")
plt.axhline(y=0.015, color='orange', linestyle='--', label="Marginally acceptable conformity")
plt.axhline(y=0.015, color='red', linestyle='-', label="Nonconformity")

plt.xticks(rotation=90)

plt.ylabel('MAD Value')
plt.title('MAD Values for each file in 10 to 10')
plt.legend()
plt.xticks(plt.xticks()[0][::5])
plt.tight_layout()
plt.show()
```
